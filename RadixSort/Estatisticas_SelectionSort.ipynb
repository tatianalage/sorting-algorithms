# Análises Estatísticas dos Tempos de Execução

Neste notebook, realizamos uma análise estatística dos tempos de execução do algoritmo de Ordenação por Inserção em duas versões: uma não otimizada e outra otimizada.

```python
import pandas as pd

# Carrega os dados do arquivo de texto para uma lista
with open('sorting_times.txt', 'r') as file:
    lines = file.readlines()

# Separa os dados em duas listas: Unoptimized e Optimized
unoptimized_data = []
optimized_data = []
for line in lines:
    parts = line.split(',')
    unoptimized_part = parts[0].split('-')[-1].strip()
    optimized_part = parts[1].split('-')[-1].strip()
    unoptimized_data.append(int(unoptimized_part[:-3]))  # Remove o "ns" e converte para int
    optimized_data.append(int(optimized_part[:-3]))  # Remove o "ns" e converte para int

# Cria um DataFrame do pandas com as duas listas
data = pd.DataFrame({'Unoptimized': unoptimized_data, 'Optimized': optimized_data})
```

```python
# Calcula as estatísticas para o algoritmo não otimizado
mean_unoptimized = data['Unoptimized'].mean()
median_unoptimized = data['Unoptimized'].median()
variance_unoptimized = data['Unoptimized'].var()
std_deviation_unoptimized = data['Unoptimized'].std()
mode_unoptimized = data['Unoptimized'].mode()[0]

# Exibe as estatísticas para o algoritmo não otimizado
print('Estatísticas para o algoritmo Radix Sort não otimizado:')
print('Média:', mean_unoptimized, 'ns')
print('Mediana:', median_unoptimized, 'ns')
print('Variância:', variance_unoptimized, 'ns')
print('Desvio padrão:', std_deviation_unoptimized, 'ns')
print('Moda:', mode_unoptimized, 'ns')
```

```python
import matplotlib.pyplot as plt

# Criando os histogramas
plt.figure(figsize=(10, 6))

# Histograma para os tempos do algoritmo não otimizado
plt.hist(unoptimized_data, bins=20, color='blue', alpha=0.7, label='Unoptimized')
plt.xlabel('Tempo (ns)')
plt.ylabel('Frequência')
plt.title('Histograma dos Tempos do Radix Sort Não Otimizado')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Calcula as estatísticas para o algoritmo otimizado
mean_optimized = data['Optimized'].mean()
median_optimized = data['Optimized'].median()
variance_optimized = data['Optimized'].var()
std_deviation_optimized = data['Optimized'].std()
mode_optimized = data['Optimized'].mode()[0]

# Exibe as estatísticas para o algoritmo otimizado
print('\nEstatísticas para o algoritmo Selecion Sort otimizado:')
print('Média:', mean_optimized, 'ns')
print('Mediana:', median_optimized, 'ns')
print('Variância:', variance_optimized, 'ns')
print('Desvio padrão:', std_deviation_optimized, 'ns')
print('Moda:', mode_optimized, 'ns')
```

```python
# Histograma para os tempos do algoritmo otimizado
plt.figure(figsize=(10, 6))
plt.hist(optimized_data, bins=20, color='green', alpha=0.7, label='Optimized')
plt.xlabel('Tempo (ns)')
plt.ylabel('Frequência')
plt.title('Histograma dos Tempos do Radix Sort Otimizado')
plt.legend()
plt.grid(True)
plt.show()
```
